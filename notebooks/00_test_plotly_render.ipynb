```python
%load_ext autoreload
%autoreload 2

import random

from plotly.offline import init_notebook_mode

init_notebook_mode(connected=True)

import plotly.graph_objects as go

# pio.renderers.default = "notebook_connected"

# Generate random data
x = [random.gauss(0, 1) for _ in range(100)]
y = [random.gauss(0, 1) for _ in range(100)]
colors = [random.gauss(0, 1) for _ in range(100)]

fig = go.Figure(
    data=go.Scatter(
        x=x,
        y=y,
        mode="markers",
        marker=dict(size=12, color=colors, colorscale="Viridis", showscale=True),
    )
)
fig.update_layout(
    title="Random Scatter Plot with Plotly",
    xaxis_title="Random X",
    yaxis_title="Random Y",
)
# fig.show()
```
